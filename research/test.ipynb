```python
import os
import json

def run_command_call(command):
    try:
        # 设置环境变量，确保能够找到 scancode 工具
        env = os.environ.copy()
        env["PATH"] = r'c:\\Users\\hzZeng\\Desktop\\license_llm\\src\\scancode-toolkit\\scancode'
        
        # 执行命令并获取返回状态
        result = os.system(command)
        if result == 0:
            print(f"Command succeeded: {' '.join(command)}")
        else:
            print(f"Command failed: {' '.join(command)} with return code {result}")
    except Exception as e:
        print(f"Exception while running command: {e}")

def getdetectedlicensesNumber(folder_path: str):
    """
    获取文件夹内所有文件的许可证数量
    """
    filename = folder_path.split('\\')[-1]
    output_file = f"scanresult/{filename}_result.json"
    scan_options = '-l'  # 默认扫描选项
    command = f'c:\\Users\\hzZeng\\Desktop\\license_llm\\src\\scancode-toolkit\\scancode {scan_options} --license-text --json-pp {output_file} {folder_path}'
    
    print(f"Scanning folder: {folder_path}...")
    run_command_call(command)
    
    try:
        with open(output_file) as f:
            data = json.load(f)
            filedetails = data['license_detections']
            return len(filedetails)
    except Exception as e:
        print(f"Error reading JSON result: {e}")
        return 0

def scan_directory_and_count_licenses(directory: str):
    """
    扫描目录中的一级文件夹，统计每个文件夹内所有文件的许可证数量
    """
    folder_counts = {}  # 用于存储每个文件夹的许可证数量

    # 只遍历一级子文件夹
    for folder in os.listdir(directory):
        folder_path = os.path.join(directory, folder)
        if os.path.isdir(folder_path):  # 如果是目录
            print(f"\nScanning folder: {folder_path}")
            total_licenses = getdetectedlicensesNumber(folder_path)
            
            # 获取该文件夹中所有文件的数量
            total_files = len([f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
            
            folder_counts[folder] = {'total_files': total_files, 'total_licenses': total_licenses}
    
    return folder_counts

def main():
    # 设置需要扫描的根目录
    directory_to_scan = r'pypi_source'  # 请根据需要修改文件夹路径
    
    # 调用 scan_directory_and_count_licenses 方法扫描目录
    result = scan_directory_and_count_licenses(directory_to_scan)
    
    # 输出每个文件夹的文件数量和许可证数量
    for folder, counts in result.items():
        print(f"\nFolder: {folder}")
        print(f"Total files: {counts['total_files']}")
        print(f"Total licenses detected: {counts['total_licenses']}")

if __name__ == "__main__":
    main()

    
    
    
```

```python
import os
import json

def find_json_files(directory):
    # 遍历指定目录下的文件和文件夹，只处理当前目录的第一级
    list = []
    for file in os.listdir(directory):
        file_path = os.path.join(directory, file)
        # 判断是否是文件，并且扩展名为 .json
        if os.path.isfile(file_path) and file.endswith('.json'):
            list.append(file_path)
    return list

def read_json(file_path):
    # 打开并读取 JSON 文件
    with open(file_path, 'r', encoding='utf-8') as f:
        return json.load(f)
print(os.getcwd())
# 示例：遍历当前目录下的所有 JSON 文件
directory = 'scanresult'  # 可以替换成你需要遍历的目录路径
for json_file in find_json_files(directory):
    print(f"Found JSON file: {json_file}")
    data = read_json(json_file)
    licenses_detecteds = data['license_detections']
    license_count = []
    for license_detected in licenses_detecteds:
        license_name = license_detected['license_expression_spdx']
        if license_name not in license_count:
            license_count.append(license_name)
    print(len(license_count))
    print(license_count)
        
print("done")
```

```python
import os
import json
import csv

spdx2context = {}

def find_json_files(directory):
    json_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".json"):
                json_files.append(os.path.join(root, file))
    return json_files

count = 0
# 示例用法
directory = "scanresult"
json_files = find_json_files(directory)
for json_file in json_files:
    with open(json_file, 'r') as f:
        data = json.load(f)
        filedetails = data['files']
        for fileitem in filedetails:
            if fileitem['license_detections']:
                spdx_expression = fileitem['license_detections'][0]['matches'][0]['license_expression_spdx']
                matched_text = fileitem['license_detections'][0]['matches'][0]['matched_text']
                if spdx_expression not in spdx2context:
                    spdx2context[spdx_expression] = [matched_text]
                    count += 1
                else:
                    spdx2context[spdx_expression].append(matched_text)
                    count += 1

# 将字典写入 CSV 文件
with open("data.csv", 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    # 写入表头
    headers = ["SPDX License Expression"] + [f"Matched Text {i+1}" for i in range(max(len(texts) for texts in spdx2context.values()))]
    csv_writer.writerow(headers)
    # 写入数据
    for spdx_expression, matched_texts in spdx2context.items():
        row = [spdx_expression] + matched_texts
        csv_writer.writerow(row)

print(f"Total matched texts: {count}")
```
