```python
import re
import sys

with open('.\QueryProcessor.py', 'r', encoding='utf-8') as f:
            code = f.read()

import_pattern = re.compile(r"^\s*(?:import|from)\s+([\w\.]+)", re.MULTILINE)
imports = import_pattern.findall(code)
# module_directory = '.\QueryProcessor.py'

# # 将自建模块目录添加到 sys.path
# if module_directory not in sys.path:
#     sys.path.append(module_directory)
        # 去重保留唯一的模块名
modules = set(imports)

print(f"检测到以下模块：{modules}\n")
for module in modules:
    try:
        __import__(module)
        print(f"✅ 模块 '{module}' 已成功导入。")
    except ImportError as e:
        print(f"❌ 模块 '{module}' 导入失败。尝试安装...")
        

for module_name, module in sorted(sys.modules.items()):
        # 获取模块的文件路径，如果模块为内置模块，则显示为 None
        if(module_name in modules):
            module_path = getattr(module, '__file__', '内置模块')
            print(f"{module_name} {module_path}")
```

```python
import json
output_file = 'C:/Users/hzZeng/Desktop/license_llm/sample.json'
filename2license={}
filename2matchtext={}
with open(output_file) as f:
            
            data=json.load(f)
            filedetails= data['files']
            for fileitem in filedetails:
                if fileitem['license_detections']:
                    
                    filename2license[fileitem['path']]=fileitem['license_detections'][0]['matches'][0]['license_expression_spdx']
                    filename2matchtext[fileitem['path']]=fileitem['license_detections'][0]['matches'][0]['matched_text']
                    
                else :
                    filename2license[fileitem['path']]='None'
                    filename2matchtext[fileitem['path']]='None'
        
print(filename2license) 
print(filename2matchtext)
```
